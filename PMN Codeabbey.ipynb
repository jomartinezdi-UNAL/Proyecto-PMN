Sum in Loop

Problem #2  

```julia
println("Ingrese la cantidad de números que desea sumar:")

num_str = readline()
num = parse(Int64, num_str) # Parse num to an integer

println("Ingrese varios números separados por espacios (ej. 10 20 30):")
valores = readline()

# Divide la cadena en partes usando el espacio como delimitador
partes_string = split(valores, ' ')

# Convierte cada parte a un número entero (Int64)
numeros = parse.(Int64, partes_string)

# Initialize sum
total = 0

for i in 1:num
    total += numeros[i]
end

println("La suma de los números es: ", total)
```

```julia
println("Ingrese la cantidad de par de números que desea sumar:")

num_str = readline()
num = parse(Int64, num_str)

input_lines = Vector{String}(undef, num) # To store raw input lines
parsed_pairs = Vector{Vector{Int64}}(undef, num) # To store parsed integer pairs
sums = Vector{Int64}(undef, num) # To store the sum of each pair

for i in 1:num
    local_string_parts = String[] # To hold the parts for the current iteration
    local_input_line = ""

    valid_for_current_pair = false
    while !valid_for_current_pair
        println("Ingrese el par de números separado por espacios (ej. 10 20) para el par #$i:")
        local_input_line = readline()
        local_string_parts = split(local_input_line, ' ')

        if length(local_string_parts) != 2
            println("Error: Debe ingresar exactamente dos números separados por espacios. Inténtelo de nuevo.")
        else
            valid_for_current_pair = true
        end
    end

    input_lines[i] = local_input_line # Store the final valid input line
    # Convierte cada parte a un número entero (Int64)
    parsed_pairs[i] = parse.(Int64, local_string_parts)
    sums[i] = parsed_pairs[i][1] + parsed_pairs[i][2]
end

println("Las sumas de los pares de números son:")
for i in 1:num
    println("Par #$i: $(parsed_pairs[i][1]) + $(parsed_pairs[i][2]) = $(sums[i])")
end

for i in 1:num
    println("$(sums[i])")
end
```

Problem #15

```julia
println("Ingrese los 300 números para encontrar el menor y mayor valor:")

local_string_parts = String[]
valid_for_current_value = false

while !valid_for_current_value

input_line = readline()
local_string_parts = split(input_line, ' ')

   if length(local_string_parts) != 300
      println("Error: Debe ingresar exactamente 300 números separados por espacios. Inténtelo de nuevo.")
   else
      valid_for_current_value = true
   end
end

values = parse.(Int64, local_string_parts)

maximum = values[1]
min = values[1]

for i in 1:299
      if maximum < values[i+1]
      maximum = values[i+1]
      end

      if min > values[i+1]
      min = values[i+1]
      continue
      end
end

println("Los números mayores y menores son:")
print("$(maximum) $(min)")
```
