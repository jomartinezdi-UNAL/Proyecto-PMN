```julia
flag=0
count=0
cuenta_cobro=0.0

while flag==0 && count<5
count = count + 1
cuenta_cobro=count*10

x=rand()
y=floor(Int64,100*x)
entrada = readline()
numero = parse(Int64, entrada)

println("Numero aleatorio generado: ",y,".... Numero del usuario: ", numero)

if numero == y
println("Felicitaciones")
flag=1

elseif numero > y
println("Valor alto")
println("No Felicitado")
println("Su deuda es: ", cuenta_cobro," dolares")

else
println("Valor por debajo")
println("No Felicitado")
println("Su deuda es: ", cuenta_cobro," dolares")

end # This 'end' closes the 'if' statement
println("El número de intentos es:", count)

end # This 'end' closes the 'while' loop

if flag == 0
println("Ha agotado sus oportunidades, Fin del codigo")
end
```
