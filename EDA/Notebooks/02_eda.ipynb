```python
import pandas as pd

file_path = "../Data/Original.xlsx"

df = pd.read_excel(file_path)

# Eliminar las 3 primeras filas
df = df.iloc[3:].reset_index(drop=True)

# Ver resultado
df = pd.read_excel(file_path, header=3)

df.head()
```

```python
list(df.columns)
```

```python
for col in df.columns:
    print("Columna:", col)
    print(df[col].head())
    print("-" * 40)
```

```python
output_file = "../Data/reporte_columnas.txt"

with open(output_file, "w", encoding="utf-8") as f:
    for col in df.columns:
        f.write(f"Columna: {col}\n")
        f.write(df[col].head().to_string())
        f.write("\n" + "-" * 40 + "\n")

print("Reporte guardado en:", output_file)
```
