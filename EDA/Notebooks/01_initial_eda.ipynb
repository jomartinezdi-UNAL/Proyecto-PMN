```python
# Missing values per column
missing_values = df.isnull().sum()

missing_values
```

- Estos son los datos nulos por columna, vemos que algunas tienes varios datos nulos. Estos datos nulos no nos sirven para hacer cosas, entonces cuando limpieamos la tabla seguramente esos datos los eliminemos.
- ¿Esos datos nulos de que serán, hay una explicación de porqué están en nulos?

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_percentage.sort_values(ascending=False)
```
